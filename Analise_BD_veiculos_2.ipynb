```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("whitegrid")
sns.set_color_codes("muted")
g=sns.catplot(y="brand", data=df, kind="count",orient='h', height=6, aspect=1.5)
g.ax.invert_yaxis()  # labels read top-to-bottom
g.ax.xaxis.set_label_text("Marca",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Total de Veículos Para Venda", fontdict= {'size':16})
g.ax.set_title("Contagem total de veículos à venda conforme a marca",fontdict= {'size':18})
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
fig = sns.catplot(x='vehicleType',y= 'price',kind ='bar',hue="gearbox", data=df,palette = "bright", height=6, aspect=1.5)
fig.ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':16})
fig.ax.yaxis.set_label_text("Preço Médio", fontdict= {'size':16})
fig.ax.set_title("Preço Médio conforme o tipo de veículo e modelo de câmbio",fontdict= {'size':18})
plt.show()
```
