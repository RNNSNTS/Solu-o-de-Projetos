```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x='fuelType', y="price",hue="gearbox", palette="bright", data=df)
ax.set_title("Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio",fontdict= {'size':15})
ax.xaxis.set_label_text("Tipo de combustível",fontdict= {'size':12})
ax.yaxis.set_label_text("Preço Médio",fontdict= {'size':12})
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#4879C8","#EAF02B","#969588"]
sns.barplot(x='vehicleType', y='powerPS',hue="gearbox", palette=colors, data=df)
ax.set_title("Potência média de um veículo por tipo de combustível e tipo de caixa de câmbio",fontdict= {'size':15})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':12})
ax.yaxis.set_label_text("Potência Média",fontdict= {'size':12})
plt.show()
```
