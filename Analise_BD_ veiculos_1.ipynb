## Mini-Projeto 2 - Análise Exploratória em Conjunto de Dados do Kaggle

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(df['yearOfRegistration'], kde=True, alpha=1,fill=True)
ax.set_title('Distribuição de Veículos por ano de registro', fontsize= 15)
plt.ylabel("Densidade (KDE)", fontsize= 15)
plt.xlabel("Ano de Registro", fontsize= 15)
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.boxplot(x='vehicleType', y="price", data=df)
plt.xlabel("Tipo de Veículo",fontdict= {'size':14})
plt.ylabel("Range de Preço",fontdict= {'size':14})
plt.title("Análise de outliers", color = "b", fontsize= 20)
plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
sns.set_style("white")
fig, ax = plt.subplots(figsize=(8,6))
cv= sns.countplot(x="vehicleType", data=df, palette="tab10")
plt.xlabel("Tipo de Veículo",fontdict= {'size':16})
plt.ylabel("Total de veículos para venda",fontdict= {'size':16})
plt.title("Total de veículos para venda conforme o tipo do veículo", fontdict= {'size':18})
# to get the counts on the top heads of the bar
for p in cv.patches:
    cv.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+500))
```
